```
#a)	Create HR_Employee Pysaprk dataFrame. 
HR_Employee = spark.read.csv("dbfs:/FileStore/tables/HR_Employee.csv",header=True,inferSchema = True)
HR_Employee.show(truncate=False)
```

```
#b)	Show printSchema & show 10 rows of Dataset
HR_Employee.printSchema()
HR_Employee.show(10, truncate=False)
```

```
from pyspark.sql.functions import*
```

```
HR_Employee.createOrReplaceTempView("employeetable")
```

```
#c)	Most employee is working in which department
max_dept=spark.sql("select Department,Count(*) as countofemployees from employeetable group by Department order by countofemployees desc ")
max_dept.show()
```

```
#d)	Highest number of job roles
max_jobroles=spark.sql("select JobRole,Count(*) as countofjobroles from employeetable group by JobRole order by countofjobroles desc ")
max_jobroles.show()
```

```
#e)	Which gender have higher strength as workforce?
gender_Strength=spark.sql("select Gender,Count(Workex) as totalWokers from employeetable group by Gender order by totalWokers desc ")
gender_Strength.show()
```

```
#f)	Compare the marital status of employee and find the most frequent status.
marital_Status=spark.sql("select MaritalStatus,Count(*) as total from employeetable group by MaritalStatus order by total desc ")
marital_Status.show()
```

```
#g)	Mostly hired employee have qualification
education_Status=spark.sql("select Education,Count(*)as total from employeetable group by Education order by total desc ")
education_Status.show()
```

```
#h)	Find the count of employee from which education fields 
education_Field=spark.sql("select EducationField,Count(*)as total from employeetable group by EducationField order by total desc ")
education_Field.show()
```

```
#i)	What is the job satisfaction level of employee?
job_Satisfaction=spark.sql("select JobRole,JobSatisfaction,Count(JobSatisfaction)as total_satisfied from employeetable group by JobRole,JobSatisfaction order by total_satisfied desc ")
job_Satisfaction.show()
```

```
#j)	Does most of employee do overtime: Yes or No?
overtime=spark.sql("select OverTime,Count(OverTime) as total_overtime from employeetable group by OverTime order by total_overtime desc ")
overtime.show()
```

```
#k)	Find Min & Max Salaried employees.
max_salaried=spark.sql("select max(Income) from employeetable").show()
max_salaried_employee=spark.sql("select * from employeetable where Income=(select max(Income) from employeetable)").show()
min_salaried=spark.sql("select min(Income) from employeetable").show()
min_salaried_employee=spark.sql("select * from employeetable where Income=(select min(Income) from employeetable) ").show()
```

```
#l)	Does most of the employee do business travel? Find of the employees counts for each category
businessTravel_employee=spark.sql("select Department,BusinessTravel, count(BusinessTravel) as counter from employeetable group by Department,BusinessTravel order by counter desc  ").show()
```

```
# m) Find the AVG salary of graduate employee.
spark.sql("select Education, AVG(Income) as average_salary from employeetable where Education='Bachelor' group by Education").show()
```

```
# n) Find the employee qualification receiving salary lower than equal to avg. salary of all employee.
avg_sal=spark.sql("select AVG(Income) from employeetable").show()
spark.sql("select Education, AVG(Income) as avg_salary from employeetable where Income <= (select AVG(Income) from employeetable) group by Education").show()
 
```

```
# o) When does the employee have highest chance of promotion in terms of working year?
spark.sql("select * from employeetable where Performance_Rating == 'Outstanding'").show()
 
```

```
# p) Highest attrition is in which department? 
spark.sql("select Department,count(Attrition) as count_attrition from employeetable where Attrition = 'Yes'  group by Department").show()
 
```

```
# q) Show marital status of Person having highest attrition rate
spark.sql("select MaritalStatus, count(Attrition) as highest_attritionInResarch from employeetable where Attrition = 'Yes' and Department='Research & Development' group by MaritalStatus").show()
```
